## Introduction to Linear Regression

#### Linear regression models the relationship between a dependent variable, often designated y, and one or more dependent variables, often denoted x. Linear regression has many applications in multiple fields, and is used for two primary functions:<br> <br> Prediction: Regression can be used to fit a model to a data set of dependent and independent values. Once the model is created, it can then predict values of the dependent variable when given an independent variable.<br> <br> Analyze Relationships: Regression analysis can be used to quantify the strength of the relationship between the dependent and independent values.

#### The linear regression formula is commonly written as:<br> <br> $$y = \beta x + C$$<br> <br> $y$ is the independent variable, also known as the reponse, measured or endogenous variable<br> <br> $$x_1, x_2, \ldots, x_i$$<br> <br> $\beta$ is the coefficient of the $x$ variable.<br> <br> $C$ is the constant of the formula

## Goal of Linear Regression

#### The goal of linear regression is to find a line that has the lowest residuals. Residuals are the difference between the predicted $y_i$ and actual $y_p$ value. These residuals are then squared for two reasons:<br> <br>     Squaring always results in a positive number<br>     Increases the residual for outliers, thereby giving them less significance in the model<br> <br> These squared residuals are then summed all together to give us a measure of fitness of the predicted model. The lower the total summed squared residuals, the better fitted the model.<br> <br> The method of summing the squares of residuals is referred to as Ordinary Least Squares (OLS).

#### In OLS, residual error is calculated by subtracting the predicted value $y$ from the measured value $y$. We can then find the calculation for the total summed squared residual:<br> <br> $$Residual = y_i - y_p$$ $$y_p = \beta x + C$$ $$Residual = y_i - \beta x + C$$ $$Residual^2 = (y_i - \beta x - C)^2$$<br>  $Residual^2$, or $R^2$, is a measure of the goodness-of-fit for the model.

## Linear Regression using Python

```python
%matplotlib inline
```

```python
from __future__ import print_function
import statsmodels
import numpy as np
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
import statsmodels.stats.api as sms
import matplotlib.pyplot as plt
```

#### Load the Ginzberg data and read the first few lines with data.head()

```python
data = pd.read_csv("C:/Users/Aaron/Google_Drive/Resources/Datasets/Rdata/car/Ginzberg.csv")
```

```python
data.head()
```

#### For this example, we will analyze the relationship between a patient's need to see the world in black and white terms and their measure of fatalism and their rating on the Beck self-assessment scale. An algebraic representation of the below model we want to estimate:<br> <br> $$y = x_1 + x_2 + C$$<br>  Where $y = depression$, $x_1 = simplicity$, $x_2 = fatalism$ and $C$ some constant

```python
model = smf.ols('depression ~ simplicity + fatalism', data=data).fit()
```

### Interpreting the Model Summary 

#### Calling the summary method prints out a nice collection of handy stats such as $R^2$, $adjusted R^2$, the F-statistic, $p$ value and more

```python
print(model.summary())
```

#### The $p$ value is one of the most important diagnostics in the summary. Any coefficient with a large $p$ value should not be trusted. A good rule is any $p >= 0.5$ is not trustworthy.<br> <br> $R^2$, as discussed previously, is the total sum of the squares of the residuals. The adjusted $R_2$ is the $R^2$ penalized by the degrees of freedom to the number of training examples. It's typically better to rely on the adjusted $R^2$. A good $R^2$ is usually between 0.7 - 1.0

### Further Interpretation using Python and Statsmodels

#### Another great feature of Statsmodels is Partial Regression Plots, or Added Variable Plots. These allow us to look at the relationship of the dependent variable and the independent variable conditional to the other exogenous variables.

#### To find the relationship between the independent variable and the dependent variables, the residuals are calculated by regressing the independent variable compared to the independent variables excluding $X_k$.

#### The Partial Regression Plots has the slope $\beta$ and intercept zero. Using the <code>partregress_grid</code> function, we can get a quick view of the independent variable regressed against each independent variable.

```python
fig = plt.figure(figsize=(10,10))
fig = sm.graphics.plot_partregress_grid(model, fig=fig)
```

#### We can see from the lines above fatalism and simplicity do have a relationship to a patient's depression as scored by the Beck self-assessment.  Judging a little closer, it appears fatalism may be more related to depression than simplicity.

#### To explore that thought further, we can use the <code>plot_regress_exog</code> function to display a 2x2 plot containing four plots:<br> <br>     Dependent variable and fitted values with confidence intervals vs. the independent variable chosen<br>     Residuals of the model vs. the chosen independent variable<br>     Partial regression plot<br>     CCPR plot<br> <br> A CCPR plot, or Component-Component plus Residual Plot, is a way to view the effect of onein dependent variable on the dependent variable taking into account other independent variables.<br> <br> If $X_t$ is highly correlated with any of the other independent variables, it might result in underestimation of the true variance.  The Fit Plot plots the fitted values versus a chosen independent variable. It shows the prediction convidence interval and optionally plots the true dependent variable.<br> <br> We will take a look at how fatalism is related to depression.

```python
fig = plt.figure(figsize=(12,12))
fig = sm.graphics.plot_regress_exog(model, "fatalism", fig=fig)
```

#### And how compare the dependent variable, depression, to the patient's need to see the world in black and white.

```python
fig = plt.figure(figsize=(12,12))
fig = sm.graphics.plot_regress_exog(model, "simplicity", fig=fig)
```

#### We can see simplicity and fatalism are plotted rather evenly through the residuals, giving more evidence depression is related to a patient's need to see the world in black and white terms and the level of fatalism the patient exhibits.
