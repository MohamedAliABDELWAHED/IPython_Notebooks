```python
from __future__ import print_function
import datetime
import numpy as np
from matplotlib.finance import quotes_historical_yahoo
from matplotlib.dates import YearLocator, MonthLocator, DateFormatter
from sklearn.hmm import GaussianHMM
import pylab as pl
import pandas as pd
```

```python
from xlwings import Range, Chart, Workbook
wb = Workbook()
```

```python
startdate = datetime.date(1990, 1, 1)
enddate = datetime.date(2012, 5, 1)
quote = quotes_historical_yahoo("GE", startdate, enddate)
if len(quote) == 0:
    raise SystemExit
```

```python
dates = np.array([i[0] for i in quote], dtype=int)
close = np.array([i[2] for i in quote])
volume = np.array([i[5] for i in quote])[1:]
```

```python
close_diff = close[1:] - close[:-1]
dates = dates[1:]
close = close[1:]
```

```python
x = np.column_stack([close_diff, volume])
```

```python
components = 5
```

```python
hmm = GaussianHMM(components, covariance_type="diag", n_iter=1500)
```

```python
hmm.fit([x])
```

```python
fit = hmm.fit([x])
```

```python
states = hmm.predict(x)
```

```python
years = YearLocator()
months = MonthLocator()
yearsFmt = DateFormatter('%Y')
```

```python
print("Transition matrix")
print(hmm.transmat_)
print()

print("means and vars of each hidden state")
for i in range(components):
    print("%dth hidden state" % i)
    print("mean = ", hmm.means_[i])
    print("var = ", np.diag(hmm.covars_[i]))
    print()

years = YearLocator()
months = MonthLocator()
yearsFmt = DateFormatter('%Y')
fig = pl.figure(figsize=(12,8))
ax = fig.add_subplot(111)

for i in range(components):
    index = (states == i)
    ax.plot_date(dates[index], close[index], 'o', label="%dth hidden state" % i)
ax.legend()

ax.xaxis.set_major_locator(years)
ax.xaxis.set_major_formatter(yearsFmt)
ax.xaxis.set_minor_locator(months)
ax.autoscale_view()

ax.fmt_xdata = DateFormatter('%Y-%m-%d')
ax.fmt_ydata = lambda x: '$%1.2f' % x
ax.grid(True)

fig.autofmt_xdate()
pl.show()
```
